```python
def get_user_choice(category, options):
    """Helper function to display a menu and return the user's selection."""
    print(f"\n--- Select your {category} ---")
    for i, option in enumerate(options, 1):
        print(f"{i}. {option}")
    print(f"{len(options) + 1}. Type my own custom {category.lower()}")
    
    try:
        choice = input(f"Enter the number (1-{len(options) + 1}) or 'quit': ")
        if choice.lower() == 'quit':
            return 'quit'
        
        choice_idx = int(choice)
        if 1 <= choice_idx <= len(options):
            return options[choice_idx - 1]
        else:
            return input(f"Enter your custom {category.lower()}: ")
    except ValueError:
        print("Invalid input. Defaulting to 'General'.")
        return "General"

profiles = ["AIML Undergrad", "Working Professional", "High School Student", "Hobbyist"]
goals = ["AI for Medical Diagnosis", "Applied Data Science", "Web Development", "Cloud Computing"]
levels = ["Beginner","Intermediate","Advanced"]
```
